```python
import requests
url = "https://wallpaper.dog/large/5439024.jpg"
r = requests.get(url, stream=True); r.raise_for_status()
with open("5439024.jpg", "wb") as f:
    for chunk in r.iter_content(8192):
        f.write(chunk)
```

resizing image using pillow library

```python
from PIL import Image
```

```python
img=Image.open('5439024.jpg')
img_resized = img.resize((150,150))
```

```python
img_resized.save('resized_image.jpg')
```

```python
img=mpimg.imread('resized_image.jpg')
img_plot = plt.imshow(img)
plt.show()
```
